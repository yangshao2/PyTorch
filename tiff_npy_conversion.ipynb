```python
import os
import numpy as np
import tifffile as tiff

def convert_tiff_to_npy(directory):
    # Iterate through all files in the directory
    for filename in os.listdir(directory):
        # Check if the file is a TIFF file
        if filename.endswith('.tiff'):
            file_path = os.path.join(directory, filename)
            # Read the TIFF file
            image = tiff.imread(file_path)
            # Define the new file name with .npy extension
            new_file_path = os.path.join(directory, filename.replace('.tiff', '.npy'))
            # Save the image data to a .npy file
            np.save(new_file_path, image)
            print(f"Converted and saved: {new_file_path}")

# Specify the directory containing the TIFF files
directory = '/home/yshao/bog/train/images'
convert_tiff_to_npy(directory)
```
